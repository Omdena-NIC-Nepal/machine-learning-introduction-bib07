```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Load dataset
data=pd.read_csv("../data/BostonHousing.csv")
data.head()
```

```python
data['b'].unique()
```

```python
#Data information
print(data.info())
```

```python
print(data.describe())
```

```python
#Chedking missing values
data.isnull().sum()
```

```python
#Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#Pairplot for selected features
selected_features = ['rm', 'lstat', 'ptratio', 'medv']
sns.pairplot(data[selected_features])
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Distribution of target variable
sns.histplot(data['medv'], bins=30, kde=True)
plt.title('Distribution of medv (Target)')
plt.xlabel('Median Value of Homes (in $1000s)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Detecting outliers using boxplots
plt.figure(figsize=(15, 6))
for i, feature in enumerate(['rm', 'lstat', 'ptratio']):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=data[feature])
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```
